# t-SNE 作業流程

```python
# 載入相關套件
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE # manifold 流形還原
```

```python
# 以iris資料集為例
iris = pd.read_csv('C:/Users/hsu/Desktop/DSLab/testpy/iris.csv')
df = iris.copy()
```

```python
# 物件
sne = TSNE()
```

```python
df.head()
```

```python
df.iloc[:,0:-1]
```

```python
# t-SNE降維
sne_data = sne.fit_transform(df.iloc[:,0:-1])
```

```python
# 先將資料目標變數數值化
df['Species'] = df['Species'].replace({'setosa':0, 'versicolor':1, 'virginica':2})
```

```python
# 將原先4個維度，降低為2個維度，繪製平面圖
plt.scatter(sne_data[:,0], sne_data[:,1], c = df['Species'])

# 透過資料視覺化可以看出約莫為三群，每一群有集中的傾向
```

# 繪製3D t-SNE圖

```python
from mpl_toolkits import mplot3d 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
sne = TSNE(n_components=3)
```

```python
sne_3d = sne.fit_transform(df.iloc[:,0:4])
```

```python
fig = plt.figure(figsize=[20,10])
ax = plt.axes(projection="3d") 
ax.scatter3D(sne_3d[:,0], sne_3d[:,1], sne_3d[:,2], c = df['Species'])

# 如要繪製3D圖上述三行需同時出現在同一個cell

plt.show()
```
